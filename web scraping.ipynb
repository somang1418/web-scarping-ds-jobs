```python
import numpy as np
import pandas as pd
import requests
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
from time import sleep
```

```python
url= 'https://www.indeed.com/jobs?q=data+scientist'
headers = {'User-Agent':'somang1418@gmail.com', 'access-to':'https://www.indeed.com/jobs?q=data+scientist','Accept': 'text/html'}

try:
    r = requests.get(url, headers=headers)
    print(r)
except: 
    requests.exceptions.ConnectionError
    r.status_code = "Connection refused"

rd_soup = BeautifulSoup(r.content, "lxml")
each_job = rd_soup.findAll('div', {"class":"title"})


company_name_list=[]
names=rd_soup.findAll('span',{'class':'company'})

for i in names:
    company_name_list.append(i.text)

r_company_name=[]
r_job_title=[]
r_description=[]

for soup_element in each_job:
    job_link=[a['href'] for a in soup_element.find_all('a', href=True)]
    job_title=[a['title'] for a in soup_element.find_all('a',title=True)]
    
    for i in job_link:
        if i.startswith('/rc/'):
            partial_url=i
            each_job_url= 'https://www.indeed.com/' + partial_url
            index=each_job.index(soup_element)
            company_name=company_name_list[index]
            r_job_title.append(job_title)
            r_company_name.append(company_name)
        
            try:
                headers = {'User-Agent':'somang1418@gmail.com', 'access-to':'https://www.indeed.com/jobs?q=data+scientist','Accept': 'text/html'}
                r = requests.get(each_job_url, headers=headers)
                print(r)
            except: 
                requests.exceptions.ConnectionError
                r.status_code = "Connection refused"
            rd_soup = BeautifulSoup(r.content, "lxml")
            each_job_description = rd_soup.find('div', {"id":"jobDescriptionText"})
            r_description.append(each_job_description.text)
   
```

```python
for page_num in [i for i in range(10,1000, 10)]: ## 100 page scraping 
    
    url= 'https://www.indeed.com/jobs?q=data+scientist&start=' + str(page_num)
    
    headers = {'User-Agent':'somang1418@gmail.com', 'access-to':'https://www.indeed.com/jobs?q=data+scientist','Accept': 'text/html'}
    
    try:
        r = requests.get(url, headers=headers)
        print(r)
    except: 
        requests.exceptions.ConnectionError
        r.status_code = "Connection refused"

    rd_soup = BeautifulSoup(r.content, "lxml")
    each_job = rd_soup.findAll('div', {"class":"title"})
    
    company_name_list=[]
    names=rd_soup.findAll('span',{'class':'company'})
    for i in names:
        company_name_list.append(i.text)
        

        
    for soup_element in each_job:
        job_link=[a['href'] for a in soup_element.find_all('a', href=True)]
        job_title=[a['title'] for a in soup_element.find_all('a',title=True)]
    
        for i in job_link:
            
            #without ad
            if i.startswith('/rc/'):
                partial_url=i
                each_job_url= 'https://www.indeed.com/' + partial_url
                index=each_job.index(soup_element)
                company_name=company_name_list[index]
                r_job_title.append(job_title)
                r_company_name.append(company_name)
        
                try:
                    headers = {'User-Agent':'somang1418@gmail.com', 'access-to':'https://www.indeed.com/jobs?q=data+scientist','Accept': 'text/html'}
                    r = requests.get(each_job_url, headers=headers)
                    print(r)
                except: 
                    requests.exceptions.ConnectionError
                    r.status_code = "Connection refused"
                rd_soup = BeautifulSoup(r.content, "lxml")
                each_job_description = rd_soup.find('div', {"id":"jobDescriptionText"})
                r_description.append(each_job_description.text)
            
                
    print("so far page:" + str(1+round(page_num/10)))
    sleep(60) #give the website some rest
   
```
